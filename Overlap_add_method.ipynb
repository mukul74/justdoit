```python
import numpy as np
sequence = np.arange(0,1000)
print(sequence)
```

```python
winsize = 64
step = 32
numOfChunks = (len(sequence) - winsize)//step + 1
print(numOfChunks)
```

```python
arr = []
for i in range(0,numOfChunks*step ,step):
    arr.append(sequence[i:i+winsize])
    # print(sequence[i:i+winsize]) 
    pass
```

```python
print(arr)
```

```python
res = np.zeros(1000)
offets = range(0,numOfChunks*step ,step)
for i, n in enumerate(offets):
    res[n:n+winsize] += arr[i]
print(res)
```
