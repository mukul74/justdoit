```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

# Simple Linear Regression Example
 
y = ax + b \|
"a" is the slope and "b" is the intercept

```python
rng = np.random.RandomState(2)
x = 10 * rng.rand(100)
y = 2 * x - 5 + rng.randn(100)
plt.scatter(x,y)
```

Using the scikit-learn's linear regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis], y)
xfit = np.linspace(0,10,1000)
yfit = model.predict(xfit[:, np.newaxis])
plt.scatter(x,y)
plt.plot(xfit, yfit)
```

Printing the model's slope and intercept

```python
print("Predicted Slope : ", model.coef_[0])
print("Predicted Intercept : ", model.intercept_)
```

We can also try to use LinearRegression Model to get regression analysis of a multidimentional dataset

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100,3)
y = 0.5 + np.dot(X,[1.5,-2.,1.])
```

```python
model.fit(X,y)
```

```python
print(model.coef_[0:])
print(model.intercept_)
```
